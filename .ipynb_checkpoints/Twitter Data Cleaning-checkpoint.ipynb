# Import modules

```python
import pandas as pd
import html
import numpy as np
import re
```

# data pre-processing

```python
data = pd.read_csv('sample.csv')
```

```python
data
```

```python
tweets = data['Tweet Content']
tweets
```

# Decode HTML Entities and Cleaning Newlines

So, far we already got the tweets but there are encoded html entities and newlines (\n) inside the tweets and we need to decode them first.

```python
clean_data = [None] * len(tweets)
for i in range (len(tweets)):
    x = tweets[i].replace("\n"," ") #cleaning newline "\n" from the tweets
    clean_data[i] = html.unescape(x)
```

```python
clean_data
```

# Cleaning Links using regex

```python
for i in range (len(clean_data)):
    clean_data[i] = re.sub(r"http\S+", "", clean_data[i])
```

```python
clean_data
```

```python
file = open('sample.txt', 'w+')
for i in range (len(clean_data)):
    file.write(clean_data[i] + "\n")
```
